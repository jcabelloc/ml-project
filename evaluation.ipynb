## Initial Setup

```python
import os
datapath = os.path.join("./")
# Import and settings for plotings figures
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# Import panda
import pandas as pd
# Create a function to load the dataset into a panda object
def load_dataset(filename, path=datapath):
    csv_path = os.path.join(path, filename)
    return pd.read_csv(csv_path)
```

## Base Scenario

### Load Metrics

```python
metrics_lr = load_dataset('metrics_lr.csv')
metrics_rf = load_dataset('metrics_rf.csv')
metrics_ada = load_dataset('metrics_ada.csv')
metrics_nn = load_dataset('metrics_nn.csv')
labels_lr = load_dataset('labels_lr.csv')
labels_rf = load_dataset('labels_rf.csv')
labels_ada = load_dataset('labels_ada.csv')
labels_nn = load_dataset('labels_nn.csv')
```

```python
from sklearn.metrics import roc_auc_score
auc_lr = roc_auc_score(labels_lr['orig'], labels_lr['prob'])
auc_rf = roc_auc_score(labels_rf['orig'], labels_rf['prob'])
auc_ada = roc_auc_score(labels_ada['orig'], labels_ada['prob'])
auc_nn = roc_auc_score(labels_nn['orig'], labels_nn['prob'])
```

### Plot Results

```python
plt.figure(figsize=(9, 6))

def plot_roc_curve(fpr, tpr, label=None, color='b', ydown = 0, auc='0.5'):
    plt.plot(fpr, tpr, linewidth=2, label=label, color=color)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.text(0.55, 0.4 - ydown, label + ': ' + auc, fontsize=14, color=color)

    
plot_roc_curve(metrics_lr['fpr'], metrics_lr['tpr'], 'Logistic Regression', 
               color='b', ydown=0, auc="{:.3f}".format(auc_lr))
plot_roc_curve(metrics_rf['fpr'], metrics_rf['tpr'], 'Random Forest', 
               color='g', ydown=0.1, auc="{:.3f}".format(auc_rf))
plot_roc_curve(metrics_ada['fpr'], metrics_ada['tpr'], 'AdaBoost', 
               color='r', ydown=0.2, auc="{:.3f}".format(auc_ada))
plot_roc_curve(metrics_nn['fpr'], metrics_nn['tpr'], 'A-NN', 
               color='y', ydown=0.3, auc="{:.3f}".format(auc_nn))
plt.xlabel('False Positive Rate', fontsize=16)
plt.ylabel('True Positive Rate', fontsize=16)
plt.title('ROC Curve - Base Scenario', fontsize=18)
plt.legend()

plt.show()
```
